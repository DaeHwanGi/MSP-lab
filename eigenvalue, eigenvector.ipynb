```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
A = np.mat("4 -5; 2 -3")
print("A : \n",A)

eigenvalue, eigenvector = np.linalg.eig(A)
print("eigenvalue : \n",eigenvalue)
print("eigenvector : \n", eigenvector)
```

### eigenvector has normarlized

```python
print("distance of eigenvector[0] : ", np.linalg.norm(eigenvector[:,0]))
print("distance of eigenvector[1] : ", np.linalg.norm(eigenvector[:,1]))
```

### $Ax = \lambda x$

```python
np.dot(A , eigenvector[:,0])
```

```python
eigenvalue[0]*eigenvector[:,0]
```

### determinant of $A-\lambda I$

```python
np.linalg.det(A - (eigenvalue[0] * np.identity(2)))
```

### permutation matrix

```python
A = np.mat("0 1; 1 0")
print("A : \n",A)
```

```python
v1 = np.array([0.4, 1.8])

plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.axis([-3,3,-3,3])
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')

v1_arrow = plt.arrow(0,0, v1[0], v1[1], head_width=0.05, head_length=0.1, color='blue')

Av1 = np.dot(A, v1.T)
Av1_arrow = plt.arrow(0,0, Av1[0,0], Av1[0,1], head_width=0.05, head_length=0.1, color='red')

eigenvalue, eigenvector = np.linalg.eig(A)
eigenvector_arrow0 = plt.arrow(0,0, eigenvector[0,0], eigenvector[1,0], head_width=0.05, head_length=0.1, color='green')
eigenvector_arrow1 = plt.arrow(0,0, eigenvector[0,1], eigenvector[1,1], head_width=0.05, head_length=0.1, color='green')

plt.legend([v1_arrow,Av1_arrow,eigenvector_arrow0], ['v1','Av1','eigenvector'])
```
