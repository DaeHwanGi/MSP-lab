```python
import requests
import json
import csv
from bs4 import BeautifulSoup
```

```python
f = open('output_v2.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(f)
```

```python
with open('category.json', encoding='utf-8') as category:    
    movie_dict = json.load(category)


for movie_ele in (movie_dict['cards']):
    code = movie_ele.get('items')[0].get('item').get('code')
    print(code)
    
    url = "https://watcha.net/comment/list?unique_id="+str(code)+"&start_index=0&count=10&type=like"
    req = requests.get(url)
    point_dict=json.loads(req.text)
    
    total_count = point_dict.get('meta').get('total_count')
    print(point_dict.get('meta').get('total_count'))
    
    for i in range(0,total_count//10):
        url2 = "https://watcha.net/comment/list?unique_id="+str(code)+"&start_index="+str(i*10)+"&count=10&type=like"
        req2 = requests.get(url2)
        point_dict=json.loads(req2.text)
    
        for ele in point_dict['data']:
            #print(ele['rating'])
            if(ele['text'] != None and ele['rating'] != None):
                wr.writerow([ele['text'],ele['rating']])
        print(i, total_count)
                
print("done")
```
